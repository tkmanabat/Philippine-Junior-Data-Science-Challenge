```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 


df = pd.read_csv("vaccination_all_tweets.csv")
df
```

```python
df.sort_values(by='date')
# Dataset contains tweets dated from Dec. 12, 2020 to March 29, 2021 (15 weeks & 3 days).
```

```python
# Transfer column 'date' to dataframe DateTime
df['datedt'] = pd.to_datetime(df['date'])
```

```python
# Create dataframes for each unit
df['year'] = df['datedt'].dt.year
df['month'] = df['datedt'].dt.month
df['day'] = df['datedt'].dt.day
df['dayofweek'] = df['datedt'].dt.dayofweek
df['hour'] = df['datedt'].dt.hour
df['minute'] = df['datedt'].dt.minute
df['dayofyear'] = df['datedt'].dt.dayofyear
df['date_only'] = df['datedt'].dt.date
```

```python
df['month'].value_counts()
# Number of Tweets each Month (with respect to number of days per month)
# Dec - 20 days, Jan - 31 days, Feb - 28 days, March - 29 days
```

```python
counts = df['month'].value_counts()
percent100 = df['month'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
pd.DataFrame({'counts': counts, 'percent': percent100})
```

```python
tweets_df = df.groupby(["date_only"])["text"].count().reset_index()
tweets_df.columns = ["date_only", "count"]
```

```python
def plot_time_variation(plottweets_df, x='date_only', y='count', hue=None, size=1, title="", is_log=False):
    f, ax = plt.subplots(1,1, figsize=(4*size,3*size))
    g = sns.lineplot(x=x, y=y, hue=hue, data=plottweets_df)
    plt.xticks(rotation=90)
    if hue:
        plt.title(f'{y} grouped by {hue} | {title}')
    else:
        plt.title(f'{y} | {title}')
    if(is_log):
        ax.set(yscale="log")
    ax.grid(color='black', linestyle='dotted', linewidth=0.75)
    plt.show() 
```

```python
plot_time_variation(tweets_df, title="Number of Tweets Every 15 Days (Dec. '20 - March '21')",size=1.5)
```
